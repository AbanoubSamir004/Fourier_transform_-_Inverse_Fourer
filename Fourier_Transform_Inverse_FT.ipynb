```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
start_time = 0
end_time = 10
sample_rate = 50
time = np.arange(start_time, end_time, 1/sample_rate)
```

```python
frequency = 4
amplitude=1
sinewave2 = amplitude*np.sin(2*np.pi*frequency*time)

complex_signal =sinewave2

plt.figure(figsize = (20, 10))
plt.plot(time, sinewave2, 'b')
plt.ylabel('Amplitude')
plt.title(("Sine wave with a frequency of "+" "+str(frequency)+" Hz and amplitude "+str(amplitude)),fontsize=16)

plt.show()
```

```python
frequency = 7
amplitude=0.5
sinewave3 = amplitude*np.sin(2*np.pi*frequency*time)

complex_signal +=sinewave3

plt.figure(figsize = (20, 10))
plt.plot(time, sinewave3, 'b')
plt.ylabel('Amplitude')
plt.title(("Sine wave with a frequency of "+" "+str(frequency)+" Hz and amplitude "+str(amplitude)),fontsize=16)

plt.show()
```

```python
plt.figure(figsize = (18, 10))
plt.plot(time, complex_signal, 'b')
plt.ylabel('Amplitude')
plt.title(("Sine wave with a multiple frequencies"),fontsize=16)

plt.show()
```

```python
def discrete_fourier_transform(complex_signal):

    F = np.zeros(complex_signal.shape,dtype=np.complex64)
    n = complex_signal.shape[0]
    for u in np.arange(n):
        for x in np.arange(n):
            F[u] += complex_signal[x]* np.exp((-1j *2* np.pi*u*x)/n)
        
    return F
```

```python
coeff = discrete_fourier_transform(complex_signal)
```

```python
N = len(coeff)
n = np.arange(N)
T = N/sample_rate
freq = n/T 
n_leftside = N//2
f_leftside = freq[:n_leftside]

amplitude_oneside =coeff[:n_leftside]/n_leftside

plt.figure(figsize = (20,10))
plt.plot(f_leftside, abs(amplitude_oneside), 'b')
plt.xlabel('Freq (Hz)')
plt.ylabel('Discrete Fourier Transform Amplitude')
plt.title(("Fourier Transform Depicting the Frequency Componenents"),fontsize=16)
plt.show()
```

```python
def discrete_fourier_transform_inverse(coeff):

    f = np.zeros(coeff.shape,dtype = 'float')
    n = len(coeff)
    for u in np.arange(n):
        for x in np.arange(n):
            f[u] +=coeff[x]* np.exp((1j *2* np.pi*u*x)/n)
        
    return f
```

```python
inverse=discrete_fourier_transform_inverse(coeff)
```

```python
plt.figure(figsize = (18, 10))
plt.plot(time, inverse, 'b')
plt.ylabel('Amplitude')
plt.title(("Sine wave with a multiple frequencies inverse"),fontsize=16)

plt.show()
```
